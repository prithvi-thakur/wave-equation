<center><h1> 2D Wave Equation: Finite Difference Method: Vectorized

```python
import numpy as np               # for arrays
import matplotlib.pyplot as plt  # for plotting
%config InlineBackend.figure_format = 'retina' # for high res inline figures
```

<h3> Parameters

```python
# Space parameters
Lx = 10       # x-dimension
Ly = 10       # y-dimension
dx = 1       # length of one element
dy = 1
Nx = int(Lx/dx) # number of x-elements (integer)
Ny = int(Ly/dy) # number of y-elements
Nel = Nx*Ny      # total number of elements

# Time parameters
T = 5         # total time
dt = 1      # timestep
Nt = int(T/dt) # number of timesteps (integer)

# Courant Stability
Cx2 = 0.5**2
Cy2 = 0.5**2
c = 0.1          # c = wave speed
```

<h3> Meshing

```python
x0 = np.linspace(0, Lx, Nx)    # x-discretization
y0 = np.linspace(0, Ly, Ny)    # y-discretization
t0 = np.linspace(0, T, Nt)     # t-discretization

xmesh, ymesh = np.meshgrid(x0, y0)
```

```python
# Objective function
u = np.zeros((Nx, Ny, Nt))
```

```python
xv = x0[:,None]
yv = y0[None,:]
```

```python
# Initial condition
def initial_condition(xmesh, ymesh):
    """ Initial condition for the wave. Gaussian
        function. Change it here."""
    
    # Gaussian parameters
    A = 20
    xo = 5; sx = 2;
    yo = 5; sy = 2;
    
    gauss = lambda x, mu, sig: ((x - mu)/2*sig)**2
    uo = A*np.exp(-(gauss(xmesh, xo, sx) + gauss(ymesh, yo, sy)))
    
    return uo
```

```python
# Initial condition
uo = initial_condition(xmesh, ymesh)
u[:,:,0] = uo
```

```python
# Plot initial gaussian pulse
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(111)

uo = initial_condition(xmesh, ymesh)

im = ax.imshow(uo, vmin=abs(uo).min(), vmax=abs(uo).max(), \
               extent = [0, Lx, 0, Ly], interpolation='bilinear')
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_title("Initial Gaussian pulse")

plt.colorbar(im)
plt.show()
```
