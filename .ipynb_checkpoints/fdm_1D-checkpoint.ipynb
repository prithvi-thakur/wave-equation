<center><h1> 1D Wave Equation: Finite Difference Method

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML
%config InlineBackend.figure_format = 'retina'
```

<h3> Parameters

```python
# Space parameters
Lx = 10       # length of the string
dxe = 0.01    # length of one element
Nx = int(Lx/dxe) # number of elements (integer)

# Time parameters
T = 10         # total time
dt = 0.01      # timestep
Nt = int(T/dt) # number of timesteps (integer)

# Courant Stability
C = 1          # c dt/dx; c = wave speed
```

<h3> Meshing

```python
xmesh = np.linspace(0, Lx, Nx)    # x-discretization
tmesh = np.linspace(0, T, Nt)  # t-discretization
```

<h3> Objective function (displacement)

```python
# displacement matrix discretized in space and time
u = np.zeros((Nt, Nx))
```

<h3> Initial condition

```python
# Gaussian Pulse
mu = 3    # mean
sig = 0.8 # standard deviation

u[0,:] = 20*np.exp(-np.power(xmesh - mu, 2.) / (2 * np.power(sig, 2.)))
u[1,:] = u[0,:]

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(xmesh, u[0,:])
ax.set_xlabel("Position")
ax.set_ylabel("Amplitude")
ax.set_title("Initial Gaussian Pulse")

plt.show()
```

<h3> Time loop

```python
t = 0
for t in range(1, Nt-1):
    
    # Inner points: central difference scheme
    for x in range(1, Nx-1):
        u[t+1,x] = 2*u[t,x] - u[t-1,x] + C*C*(u[t,x+1] - 2*u[t,x] + u[t,x-1])
            
    # Boundary conditions
    u[t+1,0] = 0; u[t+1,Lx-1] = 0
```

<h3> Plot results

```python
fig = plt.figure()
ax = plt.axes(xlim=(0, 10), ylim=(-20, 20))
ax.set_xlabel("Position")
ax.set_ylabel("Amplitude")
ax.set_title("1D wave equation")
line, = ax.plot([], [], lw=2)

def init():
    line.set_data([], [])
    return line,

def animate(i):
    x = xmesh
    y = u[i,:]
    line.set_data(x, y)
    return line,

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=Nt, interval=10, blit=True)
```
